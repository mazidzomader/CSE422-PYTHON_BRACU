```python
import random
def cross(x, y):
    if len(x) != len(y):
        return None, None
    n = len(x)

    p1 = random.randint(1, n//3)
    p2 = random.randint((n//3)  , (2*n)//3)

    x1 = x[:p1]
    x2 = x[p1:p2]
    x3 = x[p2:n]

    # print(p1, p2)
    # print(x1 , x2, x3)

    y1 = y[:p1]
    y2 = y[p1:p2]
    y3 = y[p2:n]
    # print(y1 , y2, y3)
    c1 = y1+x2+y3
    c2 = x1 + y2 + x3

    return c1, c2



# ans1, ans2 = cross("42153982", "24185671")
# print(ans1, ans2)

# parent = ["12345678",    "23456789"  ,  "34567890"  ,  "45678901"    ,"56789012"  ,  "67890123"]
parent = []
for i in range(6):
    p = input()
    parent.append(p)


s1 = random.randint(0, 5)
s2 = random.randint(0, 5)

ans1, ans2 = cross(parent[s1], parent[s2])

print(ans1, ans2)
```
