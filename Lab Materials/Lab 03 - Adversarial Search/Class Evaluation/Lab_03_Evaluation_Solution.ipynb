```python
def l03(x, y, id):
    res = 0

    for i in range(len(x)):
        temp = abs(ord(x[i])-ord(y[i]))
        # print(temp)
        res += id[i]*temp

    return -res


x = "ATOM"
y = "MOTA"

id = "2 4 2 4 1 1 8 9"

id = list(map(int, id.split()))


# print(id)


ans = l03(x, y, id[len(x):] )
print("Ans:" ,ans)
```
